# data cleansing: 
try to find minimal and maximal price

```python
Stand_1yr = RI_P.index[RI_P.iloc[:,0] == '標準一年期']
Trans_1yr = RI_P.index[RI_P.iloc[:,0] == '可轉換一年期']
Stand_3yr = RI_P.index[RI_P.iloc[:,0] == '標準三年期']
Trans_3yr = RI_P.index[RI_P.iloc[:,0] == '可轉換一年期']
```

```python
####start from here!!! 
####try to get names of all instance type!!!!!
instance = RI_P[~RI_P[0].isin(['標準一年期','可轉換一年期','標準三年期','可轉換三年期','付款選項','不預付','部分預付','全額預付'])].iloc[:,0]
instance.head()
```

```python
#想辦法找到一個可以定位"想要價錢"的公式
#年期順序: 標準一年、可轉換一年、標準三年、可轉換三年
#不預付、部分預付、全額預付
#以一年期而言
cheap_df = pd.DataFrame(columns = [0,1,2,3,4,5,6,7])
expan_df = pd.DataFrame(columns = [0,1,2,3,4,5,6,7])
for x in range(len(instance)):
    #取得每個instance名稱所在的row index
    inst_index = instance.iloc[x,0]

    #####最便宜組合:標準一年+完全預付#####
    a = RI_P.loc[inst_index:inst_index+20,0][RI_P.loc[inst_index:inst_index+20,0] == "標準一年期"].index[0]
    b = RI_P.loc[inst_index:inst_index+20,0][RI_P.loc[inst_index:inst_index+20,0] == "可轉換一年期"].index[0]
    for i in range(a,b+1):
        if RI_P.loc[i,0] == "全額預付": #找到標準~可轉換一年期之間的"不預付"價格選項
            cheap_df = cheap_df.append(RI_P.loc[i,:])
            cheap_df.loc[i,6] = RI_P.iloc[a,0]
            cheap_df.loc[i,7] = instance.iloc[x,1]

    #####最貴組合:可轉換一年+不預付#####
    c = RI_P.loc[inst_index:inst_index+20,0][RI_P.loc[inst_index:inst_index+20,0] == "可轉換一年期"].index[0]
    d = RI_P.loc[inst_index:inst_index+20,0][RI_P.loc[inst_index:inst_index+20,0] == "標準三年期"].index[0]
    for e in range(c,d+1):
        if RI_P.loc[e,0] == "不預付": #找到標準~可轉換一年期之間的"不預付"價格選項
            expan_df = expan_df.append(RI_P.loc[e,:])
            expan_df.loc[e,6] = RI_P.iloc[c,0]
            expan_df.loc[e,7] = instance.iloc[x,1]
```

```python
column_name= RI_P.iloc[2,:].tolist()
column_name.append('年期')
column_name.append('機種')

column_name
```

```python
cheap_df.columns = column_name
expan_df.columns = column_name
```
