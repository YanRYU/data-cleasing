# read in file

```python
#######import the dataset with all town ID
TC_name = pd.read_csv(r'C:\Users\Fish\Desktop\BD team\TownCityID_ansi.csv',engine = 'python',encoding= 'ansi',sep = ',')
TC_name.head()
```

```python
#######get all directory name in "人口/人口概況"
os.chdir(r'C:\Users\Fish\Desktop\BD team\Open Domain\人口')
lst = os.listdir('人口概況')
lst
```

# merge file by loop
PK: TOWN_ID

```python
##########AUTOMATION TRIAL##############

mer = TC_name
for u in range(len(lst)):
    #THE FIRST ITERARTOR-lst[]
    #######see what file under every directory
    lst2 = os.listdir('人口概況/'+lst[u])
    #######拿到 lst2中的所有".csv"結尾的檔案位置
    #######單行迴圈寫法
    print('****in directory:****'+lst2[0])
    
    g = (i for i,e in enumerate(lst2) if ".csv" in e)
    try:
        while True:'
            
            #######檔案位置 
            path = 'C:/Users/Fish/Desktop/BD team/Open Domain/人口/人口概況/'+lst[u]+'/'+lst2[next(g)]
            print(path)
            #######拿到檔案位置後就可以讀檔了~~
            df1 = pd.read_csv(path,engine='python',sep = ',')
            print('****in file:****\n')
            print(df1.head())
            ########MERGE TWO TABLES: df1 & TC_name
            mer = pd.merge(mer,df1.iloc[:,np.r_[2,4:(df1.shape[1]-1)]],how='left',on='TOWN_ID')
            print('****in MERGED file:****\n')
            print(mer.head())
            #########OUTPUT and RESTART AGAIN
            mer.to_csv('merged/merge'+str(u)+".csv",index = False, encoding = 'ansi')
            
    except StopIteration:####USE EXCEPTION TO CATCH "StopIteration"
        pass

```
